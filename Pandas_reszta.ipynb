Zadanie 5

```python
def doswiadczenie(doswiadczenie_lata):
        if doswiadczenie_lata <= 3: return 'Junior'
        elif doswiadczenie_lata >= 4 and doswiadczenie_lata <= 8: return 'Mid'
        elif doswiadczenie_lata >= 9 and doswiadczenie_lata <= 12: return 'Senior'
        elif doswiadczenie_lata >= 13: return 'Ekspert'
        
df.insert(13,'Doswiadczenie',df.doswiadczenie_lata.apply(doswiadczenie))
```

```python
df
```

Zadanie 6

```python
def srednia(x):  
        if x=='Ekspert':
            return df[df['Doswiadczenie']==('Ekspert')].rok_rozpoczecia_pracy.mean()
        elif x=='Senior':
            return df[df['Doswiadczenie']==('Senior')].rok_rozpoczecia_pracy.mean()
        elif x=='Mid':
            return df[df['Doswiadczenie']==('Mid')].rok_rozpoczecia_pracy.mean()
        elif x=='Junior':
            return df[df['Doswiadczenie']==('Junior')].rok_rozpoczecia_pracy.mean()
df=df.fillna({'rok_rozpoczecia_pracy': round(df.Doswiadczenie.apply(srednia),0)})
```

Zadanie 8

```python
def zamiana(departament_miasto_pracownik):        
        if departament_miasto_pracownik=='Software Development': return 'Software Development Team'
        if departament_miasto_pracownik=='HR': return 'Human Resources'
        if departament_miasto_pracownik=='Data Science': return 'Data Analytics'
        return departament_miasto_pracownik

df.departament_miasto_pracownik=df.departament_miasto_pracownik.apply(zamiana) 
```
